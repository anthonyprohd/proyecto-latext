```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr

def evaluar_expresion(expr):
    """
    Evalúa una expresión matemática y devuelve su valor numérico.
    """
    try:
        # Parsear la expresión
        expr_sympy = parse_expr(expr)
        # Evaluar la expresión numéricamente
        valor = float(expr_sympy.evalf())
        return valor
    except Exception as e:
        print(f"Error al evaluar la expresión '{expr}': {e}")
        return None

def graficar_elipsoide(a, b, c):
    """
    Grafica una elipsoide dada por sus semiejes a, b, c.
    """
    # Crear los ángulos theta y phi para parametrizar la superficie
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)
    theta, phi = np.meshgrid(theta, phi)

    # Parametrización de la elipsoide
    x = a * np.sin(phi) * np.cos(theta)
    y = b * np.sin(phi) * np.sin(theta)
    z = c * np.cos(phi)

    # Crear la figura y el plano 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Graficar la superficie de la elipsoide
    ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6, edgecolor='none')

    # Etiquetas de los ejes
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Título de la gráfica
    ax.set_title(f'Elipsoide: a={a}, b={b}, c={c}')

    # Configurar la proporción de los ejes para evitar distorsiones
    ax.set_box_aspect([a, b, c])  # Proporción de los ejes

    # Mostrar la gráfica
    plt.show()

def ingresar_parametros():
    """
    Solicita al usuario que ingrese los parámetros de la elipsoide.
    """
    print("\n--- Ingreso de Parámetros para la Elipsoide ---")
    print("Puedes ingresar números o expresiones matemáticas (e.g., sqrt(6))")

    a_input = input("Ingrese el semieje en X (a): ")
    b_input = input("Ingrese el semieje en Y (b): ")
    c_input = input("Ingrese el semieje en Z (c): ")

    a = evaluar_expresion(a_input)
    b = evaluar_expresion(b_input)
    c = evaluar_expresion(c_input)

    if None in (a, b, c):
        print("Se produjo un error al ingresar los parámetros. Inténtalo de nuevo.")
        return None, None, None

    return a, b, c

def main():
    print("=== Programa para Graficar Elipsoides ===")

    while True:
        # Ingresar parámetros de la elipsoide
        a, b, c = ingresar_parametros()

        if a is None or b is None or c is None:
            # Si hubo un error en el ingreso, reiniciar el ciclo
            continue

        # Graficar la elipsoide
        graficar_elipsoide(a, b, c)

        # Preguntar al usuario si desea graficar otra elipsoide
        continuar = input("\n¿Deseas graficar otra elipsoide? (s/n): ").strip().lower()
        if continuar != 's':
            print("¡Gracias por usar el programa! Hasta luego.")
            break

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr

def evaluar_expresion(expr):
    """
    Evalúa una expresión matemática y devuelve su valor numérico.
    """
    try:
        # Parsear la expresión
        expr_sympy = parse_expr(expr)
        # Evaluar la expresión numéricamente
        valor = float(expr_sympy.evalf())
        return valor
    except Exception as e:
        print(f"Error al evaluar la expresión '{expr}': {e}")
        exit(1)

def plot_sphere(r):
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)
    theta, phi = np.meshgrid(theta, phi)

    x = r * np.sin(phi) * np.cos(theta)
    y = r * np.sin(phi) * np.sin(theta)
    z = r * np.cos(phi)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Esfera: radio = {r}')
    ax.set_box_aspect([1,1,1])  # Esfera proporcionada

    plt.show()

def plot_elipsoid(a, b, c):
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)
    theta, phi = np.meshgrid(theta, phi)

    x = a * np.sin(phi) * np.cos(theta)
    y = b * np.sin(phi) * np.sin(theta)
    z = c * np.cos(phi)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Elipsoide: a={a}, b={b}, c={c}')
    ax.set_box_aspect([a, b, c])

    plt.show()

def plot_hyperboloid_one_sheet(a, b, c):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(-2, 2, 100)
    u, v = np.meshgrid(u, v)

    x = a * np.cosh(v) * np.cos(u)
    y = b * np.cosh(v) * np.sin(u)
    z = c * np.sinh(v)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='plasma', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Hiperboloide de Una Hoja')
    ax.set_box_aspect([1,1,1])

    plt.show()

def plot_hyperboloid_two_sheets(a, b, c):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0.5, 2, 100)  # Evitar v=0 para evitar singularidades
    u, v = np.meshgrid(u, v)

    # Hoja superior
    x1 = a * np.cosh(v) * np.cos(u)
    y1 = b * np.cosh(v) * np.sin(u)
    z1 = c * np.sinh(v)

    # Hoja inferior
    x2 = a * np.cosh(v) * np.cos(u)
    y2 = b * np.cosh(v) * np.sin(u)
    z2 = -c * np.sinh(v)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x1, y1, z1, cmap='coolwarm', alpha=0.6, edgecolor='none')
    ax.plot_surface(x2, y2, z2, cmap='coolwarm', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Hiperboloide de Dos Hojas')
    ax.set_box_aspect([1,1,1])

    plt.show()

def plot_paraboloid_elliptic(a, b):
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)
    x, y = np.meshgrid(x, y)
    z = a * x**2 + b * y**2

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='inferno', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Paraboloide Elíptico: z = {a}x² + {b}y²')
    ax.set_box_aspect([1,1,1])

    plt.show()

def plot_paraboloid_hyperbolic(a, b):
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)
    x, y = np.meshgrid(x, y)
    z = a * x**2 - b * y**2

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Paraboloide Hiperbólico: z = {a}x² - {b}y²')
    ax.set_box_aspect([1,1,1])

    plt.show()

def plot_cone(a, b):
    theta = np.linspace(0, 2 * np.pi, 100)
    h = 10  # Altura del cono
    z = np.linspace(0, h, 100)
    theta, z = np.meshgrid(theta, z)

    r = (z / h) * np.sqrt(a**2 + b**2)
    x = (a / np.sqrt(a**2 + b**2)) * r * np.cos(theta)
    y = (b / np.sqrt(a**2 + b**2)) * r * np.sin(theta)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='winter', alpha=0.6, edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Cono: a={a}, b={b}, altura={h}')
    ax.set_box_aspect([1,1,1])

    plt.show()

def graficar_superficie(tipo):
    if tipo == 'esfera':
        r_input = input("Ingrese el radio de la esfera (puede usar expresiones, e.g., 'sqrt(6)'): ")
        r = evaluar_expresion(r_input)
        if r <= 0:
            print("El radio debe ser un número positivo.")
            return
        plot_sphere(r)

    elif tipo == 'elipsoide':
        print("Ingrese los semiejes de la elipsoide:")
        a_input = input("Semieje en X (a) (puede usar expresiones, e.g., 'sqrt(6)'): ")
        b_input = input("Semieje en Y (b) (puede usar expresiones): ")
        c_input = input("Semieje en Z (c) (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        c = evaluar_expresion(c_input)
        if a <= 0 or b <= 0 or c <= 0:
            print("Los semiejes deben ser números positivos.")
            return
        plot_elipsoid(a, b, c)

    elif tipo == 'hiperboloide de una hoja':
        print("Ingrese los parámetros del hiperboloide de una hoja:")
        a_input = input("Constante a (puede usar expresiones): ")
        b_input = input("Constante b (puede usar expresiones): ")
        c_input = input("Constante c (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        c = evaluar_expresion(c_input)
        plot_hyperboloid_one_sheet(a, b, c)

    elif tipo == 'hiperboloide de dos hojas':
        print("Ingrese los parámetros del hiperboloide de dos hojas:")
        a_input = input("Constante a (puede usar expresiones): ")
        b_input = input("Constante b (puede usar expresiones): ")
        c_input = input("Constante c (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        c = evaluar_expresion(c_input)
        plot_hyperboloid_two_sheets(a, b, c)

    elif tipo == 'paraboloide eliptico':
        print("Ingrese los parámetros del paraboloide elíptico:")
        a_input = input("Coeficiente a (puede usar expresiones): ")
        b_input = input("Coeficiente b (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        plot_paraboloid_elliptic(a, b)

    elif tipo == 'paraboloide hiperbolico':
        print("Ingrese los parámetros del paraboloide hiperbólico:")
        a_input = input("Coeficiente a (puede usar expresiones): ")
        b_input = input("Coeficiente b (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        plot_paraboloid_hyperbolic(a, b)

    elif tipo == 'cono':
        print("Ingrese los parámetros del cono:")
        a_input = input("Constante a (puede usar expresiones): ")
        b_input = input("Constante b (puede usar expresiones): ")
        a = evaluar_expresion(a_input)
        b = evaluar_expresion(b_input)
        if a <= 0 or b <= 0:
            print("Las constantes a y b deben ser números positivos.")
            return
        plot_cone(a, b)

    else:
        print("Forma no reconocida. Por favor, elige una opción válida.")

def main():
    formas = [
        'esfera',
        'elipsoide',
        'hiperboloide de una hoja',
        'hiperboloide de dos hojas',
        'paraboloide eliptico',
        'paraboloide hiperbolico',
        'cono'
    ]

    print("Formas Geométricas Disponibles para Graficar:")
    for i, forma in enumerate(formas, 1):
        print(f"{i}. {forma.capitalize()}")

    try:
        seleccion = int(input("\nSeleccione la forma que desea graficar (ingrese el número correspondiente): "))
        if seleccion < 1 or seleccion > len(formas):
            print("Selección inválida. Por favor, elige un número válido.")
            return
        tipo_forma = formas[seleccion - 1]
        graficar_superficie(tipo_forma)
    except ValueError:
        print("Entrada inválida. Por favor, ingresa un número entero correspondiente a la forma.")

if __name__ == "__main__":
    main()
```
